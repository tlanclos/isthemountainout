```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('assets/2022-08-06-15-00-00_original.jpg', cv.IMREAD_GRAYSCALE)
template = cv.imread('assets/mountain-template.png', cv.IMREAD_GRAYSCALE)
w, h = template.shape[::-1]

res = cv.matchTemplate(img, template, cv.TM_SQDIFF_NORMED)
min_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)
top_left = min_loc
bottom_right = (top_left[0] + w, top_left[1] + h)

print(top_left, bottom_right)

plt.imshow(img[top_left[1]:bottom_right[1],top_left[0]:bottom_right[0]], cmap = 'gray')
plt.show()
```
